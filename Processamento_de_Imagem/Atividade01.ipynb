```python
#1
import numpy as np
import matplotlib.pyplot as plt

# Criar a imagem 20x20 com pixels iguais a zero
I = np.zeros((20, 20), dtype='int')

# Criar uma cópia da imagem original para modificar
I_modified = np.copy(I)

# Iluminar os pixels na 4-vizinhança do pixel (10,10)
x, y = 10, 10
I_modified[y-1, x] = 1
I_modified[y+1, x] = 1
I_modified[y, x-1] = 1
I_modified[y, x+1] = 1

# Plotar a imagem original e a modificada na mesma figura
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(I, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(I_modified, cmap='gray')
plt.title('Imagem Modificada')
plt.axis('off')

plt.show()
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt

# cria a matriz binária de dimensão 20 x 20 com todos os valores de pixels iguais a um
img = np.ones((20,20))

# escurece os pixels na vizinhança diagonal do pixel na posição (8,15)
for i in range(-1, 2):
    for j in range(-1, 2):
        if i != 0 and j != 0:
            img[8+i, 15+j] = 0

# plota a imagem original e a modificada na mesma figura
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(img, cmap='gray', vmin=0, vmax=1)
axs[0].set_title('Imagem original')
axs[1].imshow(img, cmap='gray', vmin=0, vmax=1)
axs[1].set_title('Imagem modificada')
plt.show()
```

```python
#3
import math

def d_euclidiana(p, q):
    x1, y1 = p
    x2, y2 = q
    distancia = math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
    return distancia

p = (0, 2)
q = (2, 4)

distancia = d_euclidiana(p, q)

print(distancia)
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt

# create the 20x20 image with all pixels set to zero
image = np.zeros((20, 20))

# set the coordinates of the point p
p = (10, 10)

# loop through all the pixels in the image
for i in range(20):
    for j in range(20):
        # calculate the Euclidean distance to the point p
        distance = np.sqrt((i - p[0])**2 + (j - p[1])**2)
        # if the distance is less than or equal to four, set the pixel to white (1)
        if distance <= 4:
            image[i, j] = 1

# plot the resulting image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#6
import numpy as np
import matplotlib.pyplot as plt

def d_chebyshev(p, q):
    """
    Calculates the Chebyshev distance between two points p and q.
    """
    return max(abs(p[0] - q[0]), abs(p[1] - q[1]))

# create the 1000x1000 binary image with all pixels set to zero
image = np.zeros((1000, 1000))

# set the coordinates of the point p
p = (500, 500)

# loop through all the pixels in the image
for i in range(1000):
    for j in range(1000):
        # calculate the Chebyshev distance to the point p
        distance = d_chebyshev((i, j), p)
        # if the distance is less than or equal to 60, set the pixel to white (1)
        if distance <= 60:
            image[i, j] = 1

# plot the resulting image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#7
import numpy as np

def normaliza(I, espaco_cor):
    fmax = float(np.max(I))
    fmin = float(np.min(I))
    l, c = np.shape(I)
    for i in range(l):
        for j in range(c):
            I[i,j] = (espaco_cor[1]/(fmax - fmin))*(I[i,j]-fmin)

A = np.array([[300, 320, 330], [45, 105, 170], [255, 350, 120]])
print('Imagem A:')
print(A)

normaliza(A, [0, 255])

print('Imagem A normalizada:')
print(A)
```

```python
#10/A
import numpy as np
import matplotlib.pyplot as plt

def d_chebyshev(p, q):
    return max(abs(p[0]-q[0]), abs(p[1]-q[1]))

def d_euclidiana(p, q):
    return np.sqrt((p[0]-q[0])**2 + (p[1]-q[1])**2)

Iq = 0.5*np.ones((500,500), dtype='float')
Ic = 0.5*np.ones((500,500), dtype='float')
p = (200,200)
pc = (300,300)

for i in range(500):
    for j in range(500):
        q = (i,j)
        if d_chebyshev(p,q)<=100:
            Iq[i,j] = 0
        if d_euclidiana(pc,q)<=100:
            Ic[i,j] = 0

Isub = Iq - Ic

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(Iq, cmap='gray', vmin=0, vmax=1)
plt.title('Iq')

plt.subplot(1,3,2)
plt.imshow(Ic, cmap='gray', vmin=0, vmax=1)
plt.title('Ic')

plt.subplot(1,3,3)
plt.imshow(Isub, cmap='gray', vmin=-1, vmax=1)
plt.title('Iq - Ic')

plt.show()
```
