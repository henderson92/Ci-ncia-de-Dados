```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 2*np.pi, 500)
y = np.cos(4*t)
y1 = np.sin(4*t)
```

```python
plt.figure(1)
plt.plot(t, y)
plt.title("Grafico de Cosseno")
plt.xlabel("Tempo(S)")
plt.ylabel("Amplitude")

plt.figure(2)
plt.plot(t, y1)
plt.title("Grafico de Seno")
plt.xlabel("Tempo(S)")
plt.ylabel("Amplitude")

plt.show()
```

```python
t = np.arange(0, 101)
y = np.exp(t)
plt.figure("Grafico", figsize= (6, 4)) 
plt.plot(t, y)
plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 101)
y = np.cos(t)
y1 = np.sin(t)

plt.figure("Grafico", figsize= (6, 4)) 
plt.plot(t, y)
plt.plot(t, y1)
plt.grid()
plt.title("Gráficos do Seno e Cosseno")
plt.xlabel("Eixo de Tempo")
plt.ylabel("Eixo de Amplitude")

plt.show()
```
