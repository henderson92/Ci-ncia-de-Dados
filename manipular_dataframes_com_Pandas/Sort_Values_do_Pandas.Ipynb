Como Utilizar a função sort_values()

```python
import numpy as np
import pandas as pd
```

```python
carros = pd.DataFrame(np.random.randint(3000, 25000, size=(8,12)),
                      ['Fiat', 'Pegeout', 'Volkswagem', 'Ford', 'Nissan', 'Toyota', 'Honda', 'Chevrolet'],
                      ['Jav', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],)
carros
```

```python
carros['Total'] = carros.sum(axis = 1)
carros
```

```python
carros.sort_values(by=('Total'))
```

```python
carros['Origem'] = ['Italia', 'França', 'Alemanha', 'EUA', 'Japão', 'Japão', 'Japão', 'EUA']
carros
```

```python
carros.sort_values(by=('Total'))
```

```python
carros.sort_values(by=('Total'), ascending=False)
```

```python
top3 = carros.sort_values(by=('Total'), ascending=False).head(3)
top3
```

```python
top_piores = carros.sort_values(by=('Total'), ascending=False).tail(3)
top_piores
```
